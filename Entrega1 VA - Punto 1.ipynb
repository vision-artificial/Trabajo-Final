```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = np.zeros((400,400,3))  

for x in range(400):

    for y in range(400):
        
        if  0 < y < 100:
            img[x,y] = (255,255,0)
        elif 100 <= y < 200:
            img[x,y] = (0,255,255)
        elif 200 <= y < 300:
            img[x,y] = (255,0,0)
        else:
            img[x,y] = (0,255,0)
img
plt.imshow(img)
cv2.imwrite("res/MyImage.jpg",img)
```

```python
img = cv2.imread("res/MyImage.jpg", cv2.IMREAD_COLOR)
img_rgb  = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_R = img_rgb[:,:,0]
img_G = img_rgb[:,:,1]
img_B = img_rgb[:,:,2]
img_lab  = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
img_L = img_lab[:,:,0]
img_A =  img_lab[:,:,1]
img_B = img_lab[:,:,2]
fig, arreglo_plots = plt.subplots(6,2, figsize=(15, 25))
colors = ('r','g','b')
#SISTEMA RGB
#Canal R
arreglo_plots[0,0].set_title('Canal R')
arreglo_plots[0,0].imshow(img_R, cmap="gray")
arreglo_plots[0,1].set_title('Histograma Canal R')
img_array_i = img_rgb[:,:,0].ravel();
arreglo_plots[0,1].hist(img_array_i,histtype='step',bins=255,range=(0.0, 255.0),normed=True,color=colors[0])

#Canal G
arreglo_plots[1,0].set_title('Canal G')
arreglo_plots[1,0].imshow(img_G, cmap="gray")
arreglo_plots[1,1].set_title('Histograma Canal G')
img_array_i = img_rgb[:,:,1].ravel();
arreglo_plots[1,1].hist(img_array_i,histtype='step',bins=255,range=(0.0, 255.0),normed=True,color=colors[1])

#Canal B
arreglo_plots[2,0].set_title('Canal B')
arreglo_plots[2,0].imshow(img_B, cmap="gray")
arreglo_plots[2,1].set_title('Histograma Canal B')
img_array_i = img_rgb[:,:,2].ravel();
arreglo_plots[2,1].hist(img_array_i,histtype='step',bins=255,range=(0.0, 255.0),normed=True,color=colors[2])

#SISTEMA LAB
#Canal L
arreglo_plots[3,0].set_title('Canal L')
arreglo_plots[3,0].imshow(img_L, cmap="gray")
arreglo_plots[3,1].set_title('Histograma Canal L')
img_array_i = img_lab[:,:,0].ravel();
arreglo_plots[3,1].hist(img_array_i,histtype='step',bins=255,range=(0.0, 255.0),normed=True,color=colors[0])

#Canal A
arreglo_plots[4,0].set_title('Canal A')
arreglo_plots[4,0].imshow(img_A, cmap="gray")
arreglo_plots[4,1].set_title('Histograma Canal A')
img_array_i = img_lab[:,:,1].ravel();
arreglo_plots[4,1].hist(img_array_i,histtype='step',bins=255,range=(0.0, 255.0),normed=True,color=colors[1])

#Canal B
arreglo_plots[5,0].set_title('Canal B')
arreglo_plots[5,0].imshow(img_B, cmap="gray")
arreglo_plots[5,1].set_title('Histograma Canal B')
img_array_i = img_lab[:,:,2].ravel();
arreglo_plots[5,1].hist(img_array_i,histtype='step',bins=255,range=(0.0, 255.0),normed=True,color=colors[2]);

```
