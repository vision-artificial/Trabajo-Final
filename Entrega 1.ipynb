```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("res/muestra.jpg", cv2.IMREAD_COLOR)
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fig = plt.figure(figsize=(10,12))
plt.subplot(1,1,1)
plt.imshow(img2,cmap='gray')
plt.title("Imagen Objeto")
```
