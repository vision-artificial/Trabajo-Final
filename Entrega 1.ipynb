```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("res/muestra.jpg", cv2.IMREAD_COLOR)
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fig = plt.figure(figsize=(10,12))
plt.subplot(1,1,1)
plt.imshow(img2,cmap='gray')
plt.title("Imagen Objeto")
```

**Proceso de adquisición:** La adquisición de las imágenes se efectúa en un Laboratorio en el Bloque M17 en el campus de la universidad. Para obtener las imágenes se realiza el corte del material en una lámina, lo más pulido posible, luego de tener la lámina se realiza un proceso de pulido sobre la misma, éste pulido se hace mediante un lijado con polvo de diamante y la inducción de una corriente eléctrica por medio de un electrodo; luego de esto se procede a tomar la fotografía conectando una cámara digital de 2 mpx al microscopio. La imagen obtenida se muestra en el apartado anterior
